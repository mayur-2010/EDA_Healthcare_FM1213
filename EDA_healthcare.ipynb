```python
# 1.Dataset Overview
import pandas as pd
df = pd.read_csv("healthcare_dataset.csv")

#Show first 5 rows
print("First 5 rows of dataset:")
display(df.head())

#Show last 5 rows
print("\n Last 5 rows of dataset:")
display(df.tail())

#Shape of dataset
print("\n Shape of dataset:")
print(df.shape)

#List all column names
print("\n Column Names:")
print(df.columns.tolist())

#Data types of each feature
print("\n Data Types of Columns:")
print(df.dtypes)

#Count of missing values in each column
print("\n Missing Values per Column:")
print(df.isnull().sum())

#Count of unique values in each column
print("\n Unique Values per Column:")
print(df.nunique())
```

```python
# 2.Data Quality Checks

#Find missing values
print("Missing Values in Each Column:")
print(df.isnull().sum())

#Detect duplicate rows
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

#Identify erroneous data
print("\n Rows with invalid (negative) Age values:")
display(df[df['Age'] < 0])

#Check inconsistent formatting in text columns
print("\n Sample names before cleaning:")
print(df['Name'].head(10))

print("\n Unique Gender values (to check case consistency):")
print(df['Gender'].unique())

print("\n Unique Medical Conditions (to check spelling or case issues):")
print(df['Medical Condition'].unique())
```

```python
#3. Data Cleaning

#Standardize text formatting
df['Name'] = df['Name'].str.title()
df['Gender'] = df['Gender'].str.capitalize()
df['Medical Condition'] = df['Medical Condition'].str.capitalize()
df['Doctor'] = df['Doctor'].str.title()
df['Hospital'] = df['Hospital'].str.title()
df['Insurance Provider'] = df['Insurance Provider'].str.title()
df['Admission Type'] = df['Admission Type'].str.capitalize()
df['Test Results'] = df['Test Results'].str.capitalize()

#Remove duplicate rows
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]
print(f" Removed {before - after} duplicate rows.")

# Handle missing values
df.fillna(df.mode().iloc[0], inplace=True)
print("Missing values handled successfully.")

#Verify corrections
print("\nCheck unique values after cleaning:")
print("Gender:", df['Gender'].unique())
print("Medical Condition:", df['Medical Condition'].unique())
print("Admission Type:", df['Admission Type'].unique())

#Preview cleaned data
print("\n Cleaned Data:")
display(df.head())
```

```python
#4. Descriptive Statistics

import numpy as np

#Summary statistics for numerical columns
print("Summary Statistics:")
display(df.describe())

#Calculate mean, median, mode for each numerical column
print("\n Mean of numerical columns:")
print(df.mean(numeric_only=True))

print("\n Median of numerical columns:")
print(df.median(numeric_only=True))

print("\n Mode of numerical columns:")
print(df.mode().iloc[0])

# Variance and Standard Deviation
print("\n Variance of numerical columns:")
print(df.var(numeric_only=True))

print("\n Standard Deviation of numerical columns:")
print(df.std(numeric_only=True))

#Skewness and Kurtosis
print("\n Skewness of numerical columns:")
print(df.skew(numeric_only=True))

print("\n Kurtosis of numerical columns:")
print(df.kurtosis(numeric_only=True))

# Value counts for categorical data
print("\n Value counts for categorical columns:")
for col in df.select_dtypes(include='object').columns:
    print(f"\n {col}:")
    print(df[col].value_counts().head())
```

```python
#5. Data Transformation & Encoding

from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder

#Select numerical and categorical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include='object').columns

print("Numerical Columns:", list(num_cols))
print("Categorical Columns:", list(cat_cols))

# Scaling / Normalization
minmax_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[num_cols] = minmax_scaler.fit_transform(df_minmax[num_cols])

print("\n Applied Min-Max Scaling on numerical columns.")
display(df_minmax[num_cols].head())

#Standardization
std_scaler = StandardScaler()
df_standard = df.copy()
df_standard[num_cols] = std_scaler.fit_transform(df_standard[num_cols])

print("\nApplied Standard Scaling on numerical columns.")
display(df_standard[num_cols].head())

#Encoding categorical variables

#Label Encoding
label_enc = LabelEncoder()
df_label = df.copy()
for col in cat_cols:
    df_label[col] = label_enc.fit_transform(df_label[col])

print("\n Applied Label Encoding on categorical columns.")
display(df_label.head())

#One-Hot Encoding
df_onehot = pd.get_dummies(df, columns=cat_cols, drop_first=True)
print("\n Applied One-Hot Encoding on categorical columns.")
display(df_onehot.head())
```

```python
#6.Outlier Detection & Treatment

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 1. IQR Method
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Find outliers
outliers = (df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))
print("Number of outliers in each column:")
print(outliers.sum())

# Remove outliers
df_clean = df[outliers.any(axis=1)]
print("\nShape before removing outliers:", df.shape)
print("Shape after removing outliers:", df_clean.shape)

#2. Z-Score Method
z = (df[num_cols] - df[num_cols].mean()) / df[num_cols].std()
df_z_clean = df[(np.abs(z) < 3).all(axis=1)]
print("\nShape after removing Z-score outliers:", df_z_clean.shape)

#3. Boxplot
plt.figure(figsize=(12,5))
sns.boxplot(data=df[num_cols])
plt.title("Before Removing Outliers")
plt.show()

plt.figure(figsize=(12,5))
sns.boxplot(data=df_clean[num_cols])
plt.title("After Removing Outliers (IQR Method)")
plt.show()
```

```python
# 7.Data Visualization Univariate & Bivariate Analysis

import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
plt.style.use("seaborn-v0_8")
sns.set_palette("pastel")

#Univariate Analysis

# Histogram
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Distribution of Age")
plt.show()

#Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Billing Amount'])
plt.title("Boxplot of Billing Amount")
plt.show()

#Bar chart
plt.figure(figsize=(5,4))
df['Gender'].value_counts().plot(kind='bar', color=['skyblue','lightcoral'])
plt.title("Count of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

#Bar chart
plt.figure(figsize=(5,4))
df['Admission Type'].value_counts().plot(kind='bar', color='lightgreen')
plt.title("Admission Type Distribution")
plt.xlabel("Admission Type")
plt.ylabel("Count")
plt.show()

# Bivariate Analysis

#Scatter plot: Age vs Billing Amount
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='Billing Amount', data=df, hue='Gender')
plt.title("Age vs Billing Amount by Gender")
plt.show()

#Correlation Heatmap for numeric columns
plt.figure(figsize=(6,5))
sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Pairplot-shows relationships between multiple numerical columns
sns.pairplot(df, vars=['Age', 'Billing Amount', 'Room Number'], hue='Gender')
plt.suptitle("Pairplot of Healthcare Data", y=1.02)
plt.show()

#Grouped Bar Chart – Average Billing Amount per Medical Condition and Gender
plt.figure(figsize=(8,5))
sns.barplot(x='Medical Condition', y='Billing Amount', hue='Gender', data=df, errorbar=None)
plt.title("Average Billing Amount by Medical Condition and Gender")
plt.xticks(rotation=45)
plt.show()

#Heatmap of Correlation -for multiple features
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()

```
